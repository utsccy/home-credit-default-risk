```python
import pandas as pd
import os
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
#define utilities functions

def downcast_dtypes(df):
    """
    Downsize float64/int64 with float32/int32 to save memory space
    """
    float_cols = [c for c in df if df[c].dtype == "float64"]
    int_cols =   [c for c in df if df[c].dtype in ["int64"]]

    df[float_cols] = df[float_cols].astype(np.float32)
    df[int_cols]   = df[int_cols].astype(np.int32)

    return df

def one_hot_encoder(df, nan_as_category = True):
    """
    One-hot encoding for categorical columns with get_dummies
    """
    original_columns = list(df.columns)
    categorical_columns = [col for col in df.columns if df[col].dtype == 'object']
    df = pd.get_dummies(df, columns= categorical_columns, dummy_na= nan_as_category)
    new_columns = [c for c in df.columns if c not in original_columns]
    return df, new_columns
```

```python
installments = pd.read_csv('../input/installments_payments.zip')
installments, categorical_columns = one_hot_encoder(installments, nan_as_category= True)
```

## Feature engineering
#### How much overpay or underpay?
1. calculate payment percentage = actual payment / installments 
2. calculate payment percentage = actual payment - installments ï¼ˆhow much overpay or underpay)

#### Did the client pay on time, delay or ahead of time?
1. calculate DPD = max(payment date - installment date,0) 
2. calculate DBD = max(installment date - payment date,0) 

```python
installments['AMT_PAYMENT_PERC'] = installments['AMT_PAYMENT'] / installments['AMT_INSTALMENT']
installments['AMT_PAYMENT_DIFF'] = installments['AMT_INSTALMENT'] - installments['AMT_PAYMENT']
installments['SK_DPD'] = installments['DAYS_ENTRY_PAYMENT'] - installments['DAYS_INSTALMENT']
installments['SK_DBD'] = installments['DAYS_INSTALMENT'] - installments['DAYS_ENTRY_PAYMENT']
installments['SK_DPD']=installments['SK_DPD'].apply(lambda x: max(x,0))
installments['SK_DBD']=installments['SK_DBD'].apply(lambda x: max(x,0))
installments['SK_DPD'].fillna(30, inplace=True)
installments['SK_DBD'].fillna(0, inplace=True)
```

```python
sns.set_style('whitegrid')
installments_sample = installments.sample(10000)
sns.kdeplot(np.array(installments_sample['SK_DPD']), bw=100)
sns.kdeplot(np.array(installments_sample['SK_DBD']), bw=100)
```

### Rescale DPD and DBD columns

```python
installments['SK_DPD_SCALE'] = (installments['SK_DPD']/365.25).apply(np.exp)
installments['SK_DBD_SCALE'] = (installments['SK_DBD']/365.25).apply(np.exp)
```

### Calculate Last time (in days) client paid late/underpaid

```python
installments_last_late_in_days = installments[installments['DAYS_INSTALMENT'] < installments['DAYS_ENTRY_PAYMENT']]\
.groupby(['curr_app_id'])['DAYS_INSTALMENT'].max().reset_index()
installments_last_late_in_days.rename(columns={'DAYS_INSTALMENT':'INST_DAYS_LAST_LATE'},inplace=True)
installments_last_late_in_days.head()
```

```python
installments_last_underpaid_in_days = installments[installments['AMT_INSTALMENT'] < installments['AMT_PAYMENT']]\
.groupby(['curr_app_id'])['DAYS_INSTALMENT'].max().reset_index()
installments_last_underpaid_in_days.rename(columns={'DAYS_INSTALMENT':'INST_DAYS_LAST_UNDERPAID'},inplace=True)
installments_last_underpaid_in_days.head()
```

### Add aggregated Features

```python
aggregations = {
    'NUM_INSTALMENT_VERSION': ['nunique'],
    'AMT_INSTALMENT': ['max', 'mean', 'sum'],
    'AMT_PAYMENT': ['min', 'max', 'mean', 'sum'],
    'AMT_PAYMENT_PERC': [ 'mean',  'var'],
    'AMT_PAYMENT_DIFF': [ 'mean', 'var'],
    'SK_DPD':['max','var'],
    'SK_DBD':['max','var'],
    'SK_DPD_SCALE':['sum','max'],
    'SK_DBD_SCALE':['sum','max']
}
```

```python
installments_agg = installments.groupby('curr_app_id').agg(aggregations)
installments_agg.columns = ['INST_' + col[0] + "_" + col[1].upper() for col in installments_agg.columns.tolist()]
installments_agg=installments_agg.reset_index()
```

```python
installments_agg.head()
```

### Add installments count

```python
installments_agg['INST_COUNT'] = installments[['curr_app_id', 'prev_app_id']].groupby('curr_app_id').count().values
installments_agg['INST_COUNT_LOAN'] = installments[['curr_app_id', 'prev_app_id']].groupby('curr_app_id')['prev_app_id'].nunique().values
```

### Merge datasets

```python
installments_agg.merge(installments_last_late_in_days,how='left')
installments_agg.merge(installments_last_underpaid_in_days,how='left')
```

```python
installments_agg.info(verbose=True,memory_usage=True,show_counts=True)
```

```python
installments_agg.to_csv('../staging/installment.csv', index=False, compression='zip')
```
